```python
from ultralytics.nn.modules.block import PSABlock
from ultralytics import YOLO
import torch

data_yaml = "/content/bottle_dataset/bottle.yaml"

# Monkey patch FIRST
def forward_noattn(self, x):
    return x + self.ffn(x) if self.add else self.ffn(x)

PSABlock.forward = forward_noattn

# Then build model
m = YOLO("yolo11n.yaml")
m.load("/content/yolo11n.pt")

# Optional sanity check before training
x = torch.randn(1, 3, 640, 640)
with torch.no_grad():
    _ = m.model(x)
print("No-attention sanity check passed.")

# Train
m.train(
    data=data_yaml,
    epochs=100,
    patience=10,
    imgsz=640,
    batch=16,
    lr0=1e-4,
    optimizer="AdamW",
    workers=2,
    amp=False,
    name="ft_bottle_psa_noattn"
)
```

The following export procedure was repeated for each final model by changing checkpoint and output filenames

```python
import os, sys, types
os.environ["ULTRALYTICS_IGNORE_REQUIREMENTS"] = "1"

# Stub LiteRT to avoid interpreter collisions when onnx2tf imports it
sys.modules['ai_edge_litert'] = types.ModuleType('ai_edge_litert')
sys.modules['ai_edge_litert.interpreter'] = types.SimpleNamespace(Interpreter=object)

import onnx2tf
import tensorflow as tf
import numpy as np
from pathlib import Path
from ultralytics import YOLO

out_dir = Path("mobile_exports")
out_dir.mkdir(exist_ok=True)

best_pt = "/content/runs/detect/ft_bottle_act_hardswish/weights/best.pt"

m = YOLO(best_pt)
m.export(format="saved_model", imgsz=640, dynamic=False, keras=True)

saved_models = list(Path("/content").rglob("saved_model.pb"))
assert len(saved_models) > 0, "No saved_model.pb found."
saved_model_dir = max(saved_models, key=lambda p: p.stat().st_mtime).parent

converter = tf.lite.TFLiteConverter.from_saved_model(str(saved_model_dir))
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_types = [tf.float16]
tflite_bytes = converter.convert()

tflite_path = out_dir / "yolo11n_act_hardswish_fp16_640.tflite"
tflite_path.write_bytes(tflite_bytes)

(out_dir / "labels.txt").write_text("\n".join(m.model.names.values()), encoding="utf-8")

interp = tf.lite.Interpreter(model_path=str(tflite_path))
interp.allocate_tensors()
inp = interp.get_input_details()[0]
interp.set_tensor(inp["index"], np.zeros(inp["shape"], dtype=inp["dtype"]))
interp.invoke()

print("Invoke OK")
print("TFLite saved at:", tflite_path.resolve())
```

```python
import tensorflow as tf, numpy as np, cv2

# Load model
interp = tf.lite.Interpreter(model_path="mobile_exports/yolo11n_act_hardswish_fp16_640.tflite")
interp.allocate_tensors()
inp = interp.get_input_details()[0]
out = interp.get_output_details()[0]

print("Input:", inp["shape"], inp["dtype"])
print("Output:", out["shape"], out["dtype"])

# Prepare an image (example)
img = cv2.imread("/content/Demo pic.png")
assert img is not None, "Failed to read image. Check the path."

# Resize based on model input (assumes NHWC [1,H,W,3])
_, H, W, C = inp["shape"]
assert C == 3, f"Expected 3 channels, got {C}"

img = cv2.resize(img, (W, H))
img = img[..., ::-1] / 255.0
img = np.expand_dims(img.astype(np.float32), 0)

# Run inference
interp.set_tensor(inp["index"], img)
interp.invoke()
pred = interp.get_tensor(out["index"])

print(pred.shape, pred[:1])
```
